```python
import requests

# URL for fetching weather data
url = "https://www.weather.com/"

# Make a GET request to fetch weather data
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Extract and print the weather data
    data = response.json()
    print("Weather Data:")
    print("Temperature: {}°C".format(data['main']['temp']))
    print("Humidity: {}%".format(data['main']['humidity']))
    print("Description: {}".format(data['weather'][0]['description']))
else:
    print("Failed to fetch weather data. Status code: {}".format(response.status_code))
```

```python
import requests

# URL of the website to fetch weather data
url = 'https://weather.com'

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the HTML data from the response
    data = response.text

    # Extract relevant weather information
    # Parsing HTML data to get the necessary weather information

    # Example: Parsing temperature data from the website
    start_index = data.find('temperature')
    end_index = data.find('class', start_index)
    temperature_text = data[start_index:end_index].split('>')[1].split('<')[0]

    # Convert temperature to Celsius
    temperature_fahrenheit = int(temperature_text.split('°')[0])
    temperature_celsius = (temperature_fahrenheit - 32) * 5.0/9.0

    # Print the weather information in Celsius
    print(f'Temperature: {round(temperature_celsius, 2)}°C')
else:
    print('Error fetching weather data. Please try again.')
```

```python
import requests
from bs4 import BeautifulSoup

# URL for the weather data
url = 'https://weather.com/weather/today/l/33.44,-94.04'

# Send a GET request to the website to fetch the content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the element containing the weather data
    weather_data = soup.find('div', class_='today_nowcard-section')
    
    # Extract the specific weather information you are interested in
    temperature = weather_data.find('div', class_='today_nowcard-temp').text
    conditions = weather_data.find('div', class_='today_nowcard-phrase').text
    humidity = weather_data.find('div', class_='today_nowcard-sidecar').find_all('span')[1].text
    
    # Print out the weather information
    print(f'Temperature: {temperature}')
    print(f'Conditions: {conditions}')
    print(f'Humidity: {humidity}')
else:
    print('Failed to fetch weather data')
```
