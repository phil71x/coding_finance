```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
prices = pd.read_csv('./data/small_portfolio.csv', index_col='date', parse_dates=True)

# Calculate daily returns in decimal form
returns = prices.pct_change().dropna()

# Descriptive statistics
descriptive_stats = returns.describe()

# Rebase the prices to start at 100
prices_rebased = prices / prices.iloc[0] * 100

# Plotting time series of rebased prices with a log scale
plt.figure(figsize=(10, 6))
plt.plot(prices_rebased)
plt.yscale('log')
plt.title('Rebased Prices Over Time (Log Scale)')
plt.legend(prices_rebased.columns)
plt.show()

# Histogram of returns
returns.hist(bins=50, sharey=True, figsize=(10, 8), range=[-0.1, 0.1])
plt.show()

# Box plot of returns
returns.boxplot(figsize=(8, 6))
plt.show()

# Correlation matrix
correlation_matrix = returns.corr()

print(descriptive_stats)
print(correlation_matrix)
```
