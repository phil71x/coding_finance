```python
FILE_NAME = "sp500.csv"
FILE_FULL_PATH = DATA_PATH / 'dc_stock_data' / FILE_NAME

benchmark = pd.read_csv(FILE_FULL_PATH, parse_dates=['date'], index_col='date')
print(benchmark.info())
```

```python
prices_concat = pd.concat([prices, benchmark],axis=1).dropna()
prices_concat
```

```python
prices_merge = pd.merge(prices, benchmark, left_index=True, right_index=True, how='outer').dropna()
prices_merge
```

```python
prices_rebased = prices_merge.div(prices_merge.iloc[0]).mul(100).round(2)
prices_rebased
```

```python
prices_rebased.iloc[:, :4].plot()

```

```python
prices_relative = prices_rebased.div(prices_rebased['SP500'], axis=0).mul(100)
prices_relative
```

```python
prices_relative['UPS'].plot()
```

```python
# Assuming your DataFrame is named prices_rebased
df = prices_rebased.iloc[:, -2:]

# Calculate returns
returns = df.pct_change()

# Separate benchmark returns
benchmark_returns = returns['SP500']

# Arithmetic method
arithmetic_relative_performance = returns.sub(benchmark_returns, axis=0)
cumulative_arithmetic = (1 + arithmetic_relative_performance).cumprod()

# Geometric method
geometric_relative_performance = (1 + returns).div(1 + benchmark_returns, axis=0) - 1
cumulative_geometric = (1 + geometric_relative_performance).cumprod()

# Rebase to 100
cumulative_arithmetic = cumulative_arithmetic * 100
cumulative_geometric = cumulative_geometric * 100

# Set the first row to 100 for all columns
cumulative_arithmetic.iloc[0] = 100
cumulative_geometric.iloc[0] = 100

# Plotting
plt.figure(figsize=(8, 5))

# Plot arithmetic method
for column in cumulative_arithmetic.columns[:-1]:  # Exclude the benchmark
    plt.plot(cumulative_arithmetic.index, cumulative_arithmetic[column], linestyle='-', label=f'{column} (Arithmetic)')

# Plot geometric method
for column in cumulative_geometric.columns[:-1]:  # Exclude the benchmark
    plt.plot(cumulative_geometric.index, cumulative_geometric[column], linestyle='--', label=f'{column} (Geometric)')

plt.title('Relative Performance of Assets vs S&P500 (Rebased to 100)')
plt.xlabel('Date')
plt.ylabel('Relative Performance')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
UPS_relative = pd.merge(cumulative_arithmetic['UPS'], cumulative_geometric['UPS'], left_index=True, right_index=True, how='outer')
UPS_relative.tail(10)
```

```python
FILE_NAME = "sp500.csv"
FILE_FULL_PATH = DATA_PATH / 'dc/stock_data' / FILE_NAME

data = pd.read_csv(FILE_FULL_PATH, parse_dates=['date'], index_col='date')
print(data.info())
```

```python
pr = data.SP500.pct_change() # period return
pr_plus_one = pr.add(1)
cumulative_return = pr_plus_one.cumprod().sub(1)
cumulative_return.mul(100).plot()
```

```python
pr = data.SP500.pct_change() # period return
pr_plus_one = pr.add(1)
pr_plus_one[0] = 100
cumulative_return = pr_plus_one.cumprod()
cumulative_return.plot(logy=True)
```
